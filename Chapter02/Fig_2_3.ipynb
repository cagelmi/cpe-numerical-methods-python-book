```python
import numpy as np
import matplotlib.pyplot as plt

# Script to generate Figure 2.3

# Function plot
plt.figure(figsize=(6, 6))


# first plot
def phi1(x):
    """Nonlinear function."""
    return (1 + 0.75 * x ** 2) ** (1 / 3)


plt.subplot(2, 2, 1)
xx = np.linspace(1, 3.2, 100)
plt.plot(xx, phi1(xx), "k-", xx, xx, 'k-.')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="k", linewidth=1)
plt.xlim([1, 3.2])
plt.ylim([-0.5, 4])
plt.title(r"0 < $\phi$' < 1")

# Iterative sequence of fixed point method
x1 = np.zeros(4)
f1 = np.zeros(4)

x1[0] = 2.9
f1[0] = phi1(x1[0])
plt.plot(x1[0], 0, "ko")  # initial estimation

# Label initial iteration: x_0
plt.text(x1[0], -0.5, r"$\;x_0$", fontsize=12, verticalalignment="bottom")

for i in range(1, 4):
    # Fixed point iteration
    x1[i] = phi1(x1[i - 1])
    f1[i] = phi1(x1[i])

    # Plotting sequence for fixed point method
    plt.plot([x1[i - 1], x1[i - 1]], [0.0, f1[i - 1]], "k:")
    plt.plot([x1[i - 1], x1[i]], [f1[i - 1], 0.0], "k:")
    plt.plot(x1[i], 0, "ko")  # root estimators

    # Label iteration: x_i
    plt.text(x1[i], -0.5, rf"$\;x_{i}$", fontsize=12, verticalalignment="bottom")

plt.legend([r" $\phi$", 'x'], loc='upper left')


# second plot
def phi2(x):
    """Nonlinear function."""
    return 1 + 0.5 * x ** 2


plt.subplot(2, 2, 2)
xx = np.linspace(0.5, 4, 100)
plt.plot(xx, phi2(xx), "k-", xx, xx, 'k-.')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="k", linewidth=1)
plt.xlim([0.5, 4])
plt.ylim([-1, 7])
plt.title(r" $\phi$' > 1")

# Iterative sequence of fixed point method
x2 = np.zeros(4)
f2 = np.zeros(4)

x2[0] = 1.
f2[0] = phi2(x2[0])
plt.plot(x2[0], 0, "ko")  # initial estimation

# Label initial iteration: x_0
plt.text(x2[0], -1, r"$\;x_0$", fontsize=12, verticalalignment="bottom")

for i in range(1, 4):
    # Fixed point iteration
    x2[i] = phi2(x2[i - 1])
    f2[i] = phi2(x2[i])

    # Plotting sequence for fixed point method
    plt.plot([x2[i - 1], x2[i - 1]], [0.0, f2[i - 1]], "k:")
    plt.plot([x2[i - 1], x2[i]], [f2[i - 1], 0.0], "k:")
    plt.plot(x2[i], 0, "ko")  # root estimators

    # Label iteration: x_i
    plt.text(x2[i], -1, rf"$\;x_{i}$", fontsize=12, verticalalignment="bottom")

plt.legend([r" $\phi$", 'x'], loc='upper left')


# third plot
def phi3(x):
    """Nonlinear function."""
    return 1 - 0.5 * x ** (1 / 2)


plt.subplot(2, 2, 3)
xx = np.linspace(0.25, 1.5, 100)
plt.plot(xx, phi3(xx), "k-", xx, xx, 'k-.')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="k", linewidth=1)
plt.xlim([0.25, 1.5])
plt.ylim([-0.5, 2])
plt.title(r"-1 < $\phi$' < 0")

# Iterative sequence of fixed point method
x3 = np.zeros(4)
f3 = np.zeros(4)

x3[0] = 1.35
f3[0] = phi3(x3[0])
plt.plot(x3[0], 0, "ko")  # initial estimation

# Label initial iteration: x_0
plt.text(x3[0], -0.3, r"$\;x_0$", fontsize=12, verticalalignment="bottom")

for i in range(1, 4):
    # Fixed point iteration
    x3[i] = phi3(x3[i - 1])
    f3[i] = phi3(x3[i])

    # Plotting sequence for fixed point method
    plt.plot([x3[i - 1], x3[i - 1]], [0.0, f3[i - 1]], "k:")
    plt.plot([x3[i - 1], x3[i]], [f3[i - 1], 0.0], "k:")
    plt.plot(x3[i], 0, "ko")  # root estimators

    # Label iteration: x_i
    plt.text(x3[i], -0.3, rf"$\;x_{i}$", fontsize=12, verticalalignment="bottom")

plt.legend([r" $\phi$", 'x'], loc='upper left')


# fourth plot
def phi4(x):
    """Nonlinear function."""
    return 1 - x ** 2


plt.subplot(2, 2, 4)
xx = np.linspace(0.2, 1.1, 100)
plt.plot(xx, phi4(xx), "k-", xx, xx, 'k-.')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color="k", linewidth=1)
plt.xlim([0.2, 1.1])
plt.ylim([-0.2, 1.])
plt.title(r" $\phi$' < -1")

# Iterative sequence of fixed point method
x4 = np.zeros(4)
f4 = np.zeros(4)

x4[0] = 0.35
f4[0] = phi4(x4[0])
plt.plot(x4[0], 0, "ko")  # initial estimation

# Label initial iteration: x_0
plt.text(x4[0], -0.15, r"$\;x_0$", fontsize=12, verticalalignment="bottom")

for i in range(1, 4):
    # Fixed point iteration
    x4[i] = phi4(x4[i - 1])
    f4[i] = phi4(x4[i])

    # Plotting sequence for fixed point method
    plt.plot([x4[i - 1], x4[i - 1]], [0.0, f4[i - 1]], "k:")
    plt.plot([x4[i - 1], x4[i]], [f4[i - 1], 0.0], "k:")
    plt.plot(x4[i], 0, "ko")  # root estimators

    # Label iteration: x_i
    plt.text(x4[i], -0.15, rf"$\;x_{i}$", fontsize=12, verticalalignment="bottom")

plt.legend([r" $\phi$", 'x'], loc='upper right')

# Save figure
plt.tight_layout()
filename = 'Figure_2_3.png'
plt.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()
```
