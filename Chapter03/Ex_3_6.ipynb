```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def f(t, y):
    return -100000*y + 100000*np.exp(-t) + np.exp(-t)

Y0 = [0]

sol = solve_ivp(
    f,
    (0.0, 10),
    Y0,
    method='LSODA',
    t_eval=np.logspace(-5, 1, 100, base=10)
)

plt.semilogx(sol.t, sol.y[0])
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid()
filename = 'Figure_3.6.png'
plt.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()
```
