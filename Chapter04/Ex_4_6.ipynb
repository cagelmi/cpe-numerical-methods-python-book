```python
import numpy as np

def sphere_it(N, phi2, Bi):
    """
    Function that solves the steady-state conduction problem
    in a solid sphere using functional iteration.
    """

    # Parameter initialization
    dr = 1.0 / N
    R = np.linspace(dr, 1.0, N)
    alpha = phi2 * dr**2

    # Right-hand-side vector (includes boundary conditions)
    b = np.zeros(N + 1)
    b[-1] = 2.0 * Bi * dr

    # Finite-difference matrix (includes boundary conditions)
    A = np.zeros((N + 1, N + 1))

    for i in range(1, N):
        A[i, i] = -2.0
        A[i, i - 1] = 1.0 - dr / R[i]
        A[i, i + 1] = 1.0 + dr / R[i]

    # Boundary condition at r = 0
    A[0, 0] = -3.0
    A[0, 1] = 4.0
    A[0, 2] = -1.0

    # Boundary condition at r = 1
    A[N, N - 2] = 1.0
    A[N, N - 1] = -4.0
    A[N, N] = 3.0 + 2.0 * Bi * dr

    # Chemical reaction term
    f = np.zeros(N + 1)

    # Initial guess for functional iteration
    c0 = np.linalg.solve(A, b)
    k = 0
    error = 1.0

    # Functional iteration loop
    while error > 1e-5 and k < 100:
        f[1:N] = alpha * c0[1:N]**2
        c = np.linalg.solve(A, f + b)
        error = np.linalg.norm(c - c0)
        c0 = c
        k += 1

    return c
```

```python
N = 10
phi2 = 1.0
Bi = 1.0

c = sphere_it(N, phi2, Bi)
print(' C0 =', c[0])
print('C(N) = ', c[N])
```

```python
import numpy as np

def cylinder_it(N, phi2, Bi):
    """
    Function that solves the steady-state conduction problem
    in a solid cilinder using functional iteration.
    """

    # Parameter initialization
    dr = 1.0 / N
    R = np.linspace(dr, 1.0, N)
    alpha = phi2 * dr**2

    # Right-hand-side vector (includes boundary conditions)
    b = np.zeros(N + 1)
    b[-1] = 2.0 * Bi * dr

    # Finite-difference matrix (includes boundary conditions)
    A = np.zeros((N + 1, N + 1))

    for i in range(1, N):
        A[i, i] = -2.0
        A[i, i - 1] = 1.0 - 0.5* dr / R[i]
        A[i, i + 1] = 1.0 + 0.5* dr / R[i]

    # Boundary condition at r = 0
    A[0, 0] = -3.0
    A[0, 1] = 4.0
    A[0, 2] = -1.0

    # Boundary condition at r = 1
    A[N, N - 2] = 1.0
    A[N, N - 1] = -4.0
    A[N, N] = 3.0 + 2.0 * Bi * dr

    # Chemical reaction term
    f = np.zeros(N + 1)

    # Initial guess for functional iteration
    c0 = np.linalg.solve(A, b)
    k = 0
    error = 1.0

    # Functional iteration loop
    while error > 1e-5 and k < 100:
        f[1:N] = alpha * c0[1:N]/(1+c0[1:N])
        c = np.linalg.solve(A, f + b)
        error = np.linalg.norm(c - c0)
        c0 = c
        k += 1

    return c
```
