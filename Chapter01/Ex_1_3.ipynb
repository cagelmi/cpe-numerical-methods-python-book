```python
import numpy as np
from scipy.linalg import lu, solve_triangular
# Input matrix A and right-hand side vector b
A = np.array([[2, 1, -3],
              [-1, 3, 2],
              [3, 1, -3]], dtype=float)
b = np.array([-1, 12, 0], dtype=float).reshape(-1, 1)

# Solution using the inverse of matrix A (not recommended numerically)
x1 = np.linalg.inv(A) @ b
r1 = A @ x1 - b
print('x1 = ',x1)
print('r1 = ',r1)

# Solution using the linear solve operator (recommended)
x2 = np.linalg.solve(A, b)
r2 = A @ x2 - b
print('x2 = ',x2)
print('r2 = ',r2)

# Solution using LU factorization
# SciPy returns P, L, U such that A = P @ L @ U
P, L, U = lu(A)
b3 = P.T @ b  # from A x = b and A = P L U => L U x = P^T b
y3 = solve_triangular(L, b3, lower=True)
x3 = solve_triangular(U, y3, lower=False)
r3 = A @ x3 - b
print('x3 = ',x3)
print('r3 = ',r3)
```
