```python
import numpy as np
import matplotlib.pyplot as plt

# Script to generate Figure 2.3

# Function plot
plt.figure(figsize=(8, 5))


# first plot
def g1(x):
    """Nonlinear function."""
    return (2*x/3 + (1+x**2/3)**(1/2)) / (1-x/3)
def g2(x):
    """Nonlinear function."""
    return (2*x/3 - (1+x**2/3)**(1/2)) / (1-x/3)

xx = np.linspace(-1, 1, 100)
plt.plot(xx, g1(xx), "k-",xx,g2(xx),"k:")

plt.xlabel(r'h$\lambda$')
#plt.ylabel("Roots $\gamma$ $\gamma$")
plt.xlim([-1, 1])
plt.grid(True)
plt.legend([
    r'$\gamma_1$',
    r'$\gamma_2$',
], loc='upper left')

filename = 'Figure_3_4.png'
plt.savefig(filename, dpi=300, bbox_inches='tight')
plt.show()
```
