```python
import numpy as np

def jacobi(A, b, tol, maxit):
    """
    Solve A x = b using the Jacobi iterative method.

    Parameters
    ----------
    A : array_like, shape (n, n)
        Coefficient matrix.
    b : array_like, shape (n,) or (n, 1)
        Right-hand side vector.
    tol : float, optional
        Convergence tolerance on the residual norm ||A x - b||_2.
    maxit : int, optional
        Maximum number of iterations.

    Returns
    -------
    X : ndarray, shape (n, 1)
        Approximate solution.
    it : int
        Number of iterations performed.
    err : float
        Final residual norm ||A X - b||_2.
    """
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float)
    if b.ndim == 1:
        b = b.reshape(-1, 1)

    # Diagonal matrix D and strictly lower/upper parts L and U
    D = np.diag(np.diag(A))
    L = np.tril(A, -1)
    U = np.triu(A, 1)

    # Iteration matrix M = -D^{-1}(L+U)
    Dinv = np.diag(1.0 / np.diag(D))
    M = -Dinv @ (L + U)

    # Check that the eigenvalues satisfy the convergence condition (spectral radius < 1)
    rho = np.max(np.abs(np.linalg.eigvals(M)))
    if rho >= 1.0:
        raise ValueError("Iteration matrix does not guarantee convergence (spectral radius >= 1).")

    # g = D^{-1} b
    g = Dinv @ b

    X = np.zeros_like(b)
    it = 0
    err = np.linalg.norm(b)

    while err > tol and it < maxit:
        X = M @ X + g
        err = np.linalg.norm(A @ X - b)
        it += 1

    return X, it, err
```

```python
A = np.array([[6, -3, 1],
              [1, -7, 1],
              [2, 1, -8]], dtype=float)
b = np.array([11, 10, -15], dtype=float).reshape(-1, 1)

x = np.zeros_like(b)
tol=1e-5
maxit=100

[x,it,err]=jacobi(A, b, tol, maxit)
print('x = ',x)
print('iter = ',it)
print(' error = ',err)
```
