# Example 1.1

Input matrix A and right-hand side vector b

```python
import numpy as np
A = np.array([[2, 1, -3],
              [-1, 3, 2],
              [3, 1, -3]], dtype=float)
b = np.array([-1, 12, 0], dtype=float).reshape(-1, 1)

x = np.zeros_like(b)
```

Augmented matrix M and Gauss-Jordan elimination

```python
M = np.hstack((A, b))  # [A | b]
```

Step 1: normalize the first row by the first pivot

```python
M[0, :] = M[0, :] / M[0, 0]
```

Step 2: create zeros below the first pivot

```python
M[1, :] = M[1, :] - M[0, :] * M[1, 0]
M[2, :] = M[2, :] - M[0, :] * M[2, 0]
```

Step 3: normalize the second row by the second pivot

```python
M[1, :] = M[1, :] / M[1, 1]
```

Step 4: create zeros above/below the second pivot (Gauss-Jordan style)

```python
M[2, :] = M[2, :] - M[1, :] * M[2, 1]
```

Back substitution stage (reduced row-echelon form gives solution in last column)

```python
x[2, 0] = M[2, 3] / M[2, 2]
x[1, 0] = M[1, 3] - M[1, 2] * x[2, 0]
x[0, 0] = M[0, 3] - M[0, 2] * x[2, 0] - M[0, 1] * x[1, 0]
```

Verification of the solution

```python
r = A @ x - b
print(' r =', r)
print(' M = ', M)
print(' x = ',x)
```
