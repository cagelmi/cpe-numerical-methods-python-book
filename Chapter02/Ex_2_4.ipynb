```python
import numpy as np
import matplotlib.pyplot as plt

# Script to generate Figure 2.4
f1 = lambda x: x - x**2 + 0.5 * x**3          # nonlinear function f1(x)
f1p = lambda x: 1 - 2*x + 1.5 * x**2          # derivative of f1(x)

# Function plot
xx = np.linspace(-2, 4, 1000)
plt.figure()
plt.plot(xx, f1(xx), 'k-')
# plt.gca().set_aspect('equal', adjustable='box')  # axis square (approx. equivalent)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='k', linewidth=1)
plt.xlim([-2, 4])

# Iterative sequence of Newton's method
# Initialize arrays to store iterates and function values
N = 4  # number of Newton iterations including initial point
x = np.zeros(N)
f = np.zeros(N)

# Initial guess
x[0]=3.5
f[0]=f1(3.5)

for i in range(1, N):
    # Newton iteration
    x[i] = x[i-1] - f[i-1] / f1p(x[i-1])
    f[i] = f1(x[i])

    # Plotting sequence for Newton's method
    # Segment from previous point to previous point on x-axis (vertical line)
    xpl = [x[i-1], x[i-1]]
    fpl = [0.0, f[i-1]]
    plt.plot(xpl, fpl, 'k:')

    # Segment from previous point to current point (diagonal line to root approximation)
    xpl = [x[i-1], x[i]]
    fpl = [f[i-1], 0.0]
    plt.plot(xpl, fpl, 'k:')

plt.show()
```
